```python
import pandas as pd
from pathlib import Path

# 1) 파일 경로
in_path = "../raw_data/데이터/주식 데이터.csv"
out_path = "../output_data/top60기업주식평균_일별지표.csv"  # 저장 경로

# 2) 데이터 불러오기
df = pd.read_csv(in_path)

# 3) 평균 낼 카테고리 정의
categories = ["High", "Low", "Volume", "Close", "Volatility"]

# 4) 카테고리별 평균 계산
avg_data = {"Date": df["Date"]}

for cat in categories:
    cols = [c for c in df.columns if c.endswith(f"_{cat}")]
    avg_data[cat] = df[cols].mean(axis=1)

# 5) 데이터프레임 생성
avg_df = pd.DataFrame(avg_data)

# 6) Date 열에서 시간 정보 제거
# 문자열 기준으로 공백 앞까지만 남기기
avg_df["Date"] = avg_df["Date"].astype(str).str.split().str[0]

# 7) 정렬 및 저장
avg_df = avg_df.sort_values("Date").reset_index(drop=True)
avg_df.to_csv(out_path, index=False, encoding="utf-8-sig")

print("저장 완료:", out_path)
print(avg_df.head())
```
