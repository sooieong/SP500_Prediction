```python
# 파일 읽기
df_tnx = pd.read_csv("../raw_data/tnx.csv")
df_env = pd.read_csv("../raw_data/external_environment_variable.csv")


# 1) 환경 변수 파일의 Date를 timezone-aware로 읽기 (utc 강제)
df_env["Date"] = pd.to_datetime(df_env["Date"], utc=True, errors="coerce")

# 2) 타임존 제거 → 날짜만 추출 가능
df_env["Date"] = df_env["Date"].dt.tz_convert("UTC").dt.tz_localize(None)

# 3) 날짜만 남기기
df_env["Date"] = df_env["Date"].dt.date
df_env["Date"] = pd.to_datetime(df_env["Date"])   # datetime 통일

# 4) tnx의 Date도 통일
df_tnx["Date"] = pd.to_datetime(df_tnx["Date"], errors="coerce")
df_tnx["Date"] = df_tnx["Date"].dt.date
df_tnx["Date"] = pd.to_datetime(df_tnx["Date"])

# 5) Outer Join
merged = pd.merge(df_env, df_tnx, on="Date", how="outer")

# 6) 날짜 정렬
merged = merged.sort_values("Date")
```

### 기준금리 추가

```python
import pandas as pd

# 1. 파일 로드
df_env = pd.read_csv('../raw_data/환경변수_TNX_병합.csv', encoding='utf-8')
df_rate = pd.read_csv('../raw_data/기준금리.csv', encoding='utf-8')

# 2. 날짜 컬럼 정리 및 기준 날짜 컬럼 생성

# SPY_주식데이터 날짜 처리 (시간 정보 및 타임존 제거 후 날짜 형식 변환)
# 변환 과정을 단계별로 나누어 Time Zone을 확실히 제거합니다.
dt_series = pd.to_datetime(df_env['Date'], utc=True) # 1. UTC 기준 datetime64[ns, UTC]
dt_series = dt_series.dt.tz_localize(None) # 2. Time Zone 정보 제거 -> datetime64[ns] (Naive)
df_env['Date'] = dt_series.dt.normalize() # 3. 시간 정보 제거 (00:00:00으로 통일)

# 다른 데이터 파일 날짜 형식 변환 (이들은 datetime64[ns] (Naive) 형태가 됩니다.)
df_rate['Date'] = pd.to_datetime(df_rate['Date'])


# 3. 데이터 병합 및 값 채우기 (선행 발표 값 포함)

def merge_and_fill(df_main, df_announce, value_col_name):
    """
    발표 날짜와 주식 데이터 날짜를 모두 기준으로 삼아 병합하고,
    선행 발표 값을 ffill로 채우는 함수
    """

    # 3-1. 발표 데이터프레임 정리
    df_announce = df_announce.rename(columns={df_announce.columns[1]: value_col_name})
    df_announce = df_announce[['Date', value_col_name]]

    # 값 컬럼의 공백 제거 및 float 변환
    if df_announce[value_col_name].dtype == 'object':
        df_announce[value_col_name] = df_announce[value_col_name].astype(str).str.replace(r'[^\d\.\-]', '', regex=True)
        df_announce[value_col_name] = pd.to_numeric(df_announce[value_col_name], errors='coerce')

    # --- 핵심 수정: 모든 날짜를 기준으로 합니다. ---
    # 주식 데이터 날짜와 발표 데이터 날짜를 모두 합쳐 정렬된 모든 날짜 시계열을 만듭니다.
    all_dates = pd.concat([df_main['Date'], df_announce['Date']]).drop_duplicates().to_frame()

    # 3-2. 발표 데이터를 모든 날짜에 left join 합니다.
    # 이렇게 하면 2022-09-30과 같은 선행 발표일도 포함됩니다.
    df_merged = pd.merge(all_dates, df_announce, on='Date', how='left')

    # 3-3. 날짜를 기준으로 정렬하고, ffill을 적용하여 선행 발표 값을 채웁니다.
    df_merged = df_merged.sort_values(by='Date').reset_index(drop=True)
    df_merged[value_col_name] = df_merged[value_col_name].ffill()

    # 3-4. 원래 주식 데이터프레임에 최종 결과 병합
    # 병합 후에는 주식 데이터에 없는 행(선행 발표일)은 자동 제외됩니다.
    df_main = pd.merge(df_main, df_merged[['Date', value_col_name]], on='Date', how='left')

    return df_main

# 병합 작업 순차적 실행
df_final = df_env.copy()

# 기준금리 병합
df_final = merge_and_fill(df_final, df_rate, '기준금리')


# 4. 결과 확인 및 저장
print(df_final.head(10))
print("\n--- 최종 데이터프레임 정보 ---")
print(df_final.info())

# 최종 결과 CSV 파일로 저장
output_file = '../raw_data/top60_기준금리추가.csv'
df_final.to_csv(output_file, index=False)
print(f"\n최종 데이터는 '{output_file}' 파일로 저장되었습니다.")
```

환경변수_통합

```python
import pandas as pd
import numpy as np

def update_cpi_by_release_date(df_final_path, df_cpi_path, df_release_path, output_path):
    print("데이터 로드 및 전처리 시작...")

    # 1. 파일 로드
    df_cpi = pd.read_csv(df_cpi_path)
    df_release = pd.read_csv(df_release_path)
    df_final = pd.read_csv(df_final_path)

    # --- CPIAUCSL 데이터 정리 ---
    df_cpi['Date'] = pd.to_datetime(df_cpi['observation_date'])
    df_cpi = df_cpi.rename(columns={'CPIAUCSL': 'New_CPIAUCSL'})
    df_cpi = df_cpi.set_index('Date')['New_CPIAUCSL'] # Series로 변환

    # --- 발표일 데이터 파싱 함수 ---
    def format_release_date(date_str, year):
        if '발표 예정' in date_str:
            return None

        # '1/12(2023)' 같은 형식 처리
        if '(' in date_str:
            month_day, yr_suffix = date_str.replace(')', '').split('(')
            try:
                year = int(yr_suffix)
            except ValueError:
                pass # 연도 파싱 오류 시 기본 연도 사용
        else:
            month_day = date_str

        try:
            month, day = map(int, month_day.split('/'))
            return pd.to_datetime(f'{year}-{month}-{day}')
        except ValueError:
            return None

    # --- 발표일 (Announcement Date)과 기준 월 CPI 값 매핑 ---
    release_dates = {}
    df_release = df_release.set_index('기준 월')

    for col in df_release.columns:
        if '발표일' in col:
            base_year = int(col.split('년')[0])

            for month_idx, month_name in enumerate(df_release.index, 1):
                date_str = df_release.loc[month_name, col]

                # 1. 발표일 (Announcement Date) 계산
                release_dt = format_release_date(date_str, base_year)
                if release_dt is None:
                    continue

                # 2. CPI의 기준일 (CPI Base Date) 계산
                cpi_key = pd.to_datetime(f'{base_year}-{month_idx:02d}-01')

                if cpi_key in df_cpi.index:
                    cpi_value = df_cpi.loc[cpi_key]
                    # '발표일': 'CPI 값' 형태로 저장
                    release_dates[release_dt] = cpi_value

    # 발표일 데이터를 날짜 순으로 정렬하여 DataFrame 생성 (CPI 변경 시점)
    sorted_release_dates = sorted(release_dates.items())
    df_change_points = pd.DataFrame(sorted_release_dates, columns=['Date', 'Updated_CPIAUCSL'])

    # --- final_merged_y_close 데이터에 CPIAUCSL 값 업데이트 (ffill 방식) ---
    print("CPI 값 업데이트 적용 중...")

    df_final['Date'] = pd.to_datetime(df_final['Date'])
    df_final = df_final.set_index('Date').sort_index()

    # 1. 초기 CPI 값 (가장 오래된 발표일 이전의 값) 설정
    first_announcement_date = df_change_points['Date'].min()
    # 첫 발표일 직전의 마지막 월별 CPI 값을 초기값으로 설정
    initial_cpi_key = df_cpi[df_cpi.index < first_announcement_date].index.max()

    if pd.isna(initial_cpi_key):
        print("경고: 초기 CPI 값을 설정할 수 없습니다. CPI 데이터의 시작 날짜를 확인하세요.")
        initial_cpi_value = np.nan
    else:
        initial_cpi_value = df_cpi.loc[initial_cpi_key]

    # 2. change_points와 df_final의 날짜를 병합하여 업데이트 데이터프레임 생성
    df_updates = df_final.copy()
    df_updates['Updated_CPIAUCSL'] = np.nan
    df_updates = df_updates.combine_first(
        df_change_points.set_index('Date')
    )

    # 3. 정렬 및 초기값 채우기
    df_updates = df_updates.sort_index()

    # 첫 번째 발표일 이전에 있는 날짜에 초기 CPI 값 적용
    df_updates.loc[df_updates.index < first_announcement_date, 'Updated_CPIAUCSL'] = initial_cpi_value

    # 4. ffill(Forward Fill)을 사용하여 발표일 이후의 모든 날짜에 값을 채움
    df_updates['Updated_CPIAUCSL'] = df_updates['Updated_CPIAUCSL'].ffill()

    # 5. 원래 df_final의 날짜만 남기고 CPIAUCSL 컬럼 업데이트
    # 새로운 컬럼으로 추가
    df_final['CPIAUCSL'] = df_updates.loc[df_final.index, 'Updated_CPIAUCSL']
    df_final = df_final.reset_index() # 'Date' 컬럼 복원

    # 6. 결과 저장
    df_final.to_csv(output_path, index=False)
    print(f"CPIAUCSL 값이 업데이트된 파일이 '{output_path}'로 저장되었습니다.")
    return df_final

# --- 코드 사용 예시 ---

# # 사용자님의 파일 이름을 그대로 사용하여 함수 실행
updated_df = update_cpi_by_release_date(
    df_final_path='../raw_data/top60_기준금리추가.csv',
    df_cpi_path='../raw_data/CPIAUCSL.csv',
    df_release_path='../raw_data/-2022-2023-2024-2025.csv',
    output_path='../output_data/환경변수_통합.csv'
)
# print("\n업데이트된 DataFrame의 일부 (선택 사항):")
print(updated_df[['Date', 'CPIAUCSL']].head())
print(updated_df[['Date', 'CPIAUCSL']].tail())
```
